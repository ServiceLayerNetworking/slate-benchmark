```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
df
```

```python
plt.figure(figsize=(4.5, 4))
plt.xlabel('Latency (ms)', fontsize=18)
plt.xticks(fontsize=14, rotation=-45)
plt.yticks(fontsize=14)
plt.grid(True)
# handles, labels = plt.gca().get_legend_handles_labels()
# order = [1,2,0]
# plt.legend([handles[idx] for idx in order],[labels[idx] for idx in order], fontsize=12, loc='lower right')
plt.ylim(0, 1)
plt.xlim(left=0, right=max(df['SLATE'].max(), df['WATERFALL'].max()))
plt.plot(df['SLATE'], df['pct'], label='SLATE')
plt.plot(df['WATERFALL'], df['pct'], label='WATERFALL')
plt.legend(loc='lower right', fontsize=12)
plt.tight_layout()
plt.savefig('routing_rules_cdf.pdf')
print(f"Output PDF saved to: routing_rules_cdf.pdf")
plt.show()
```
