```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('threshold_over_rps.csv')
df_5ms = df[df['network-latency'] == 5]
df_25ms = df[df['network-latency'] == 25]
df_50ms = df[df['network-latency'] == 50]
straight_line = [i for i in range(0, 1001, 100)]
plt.figure(figsize=(6, 5))
plt.plot(straight_line, straight_line, color="red", alpha=0.5, linestyle='--')
plt.plot(df_5ms['west-rps'], df_5ms['threshold_on_west'], label='Network latency 5ms', marker='v')
plt.plot(df_25ms['west-rps'], df_25ms['threshold_on_west'], label='Network latency 25ms', marker='x')
plt.plot(df_50ms['west-rps'], df_50ms['threshold_on_west'], label='Network latency 50ms', marker='o')
plt.grid(True)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.ylim(bottom=0)
# plt.ylim(top=550)
plt.xlim(left=0)

plt.xlim(right=1000)
plt.ylim(top=1000)
plt.xlabel('Load on west cluster (req/sec)', fontsize=18)
plt.ylabel('Locally served load', fontsize=18)
plt.legend(loc="lower right", fontsize=12)
plt.tight_layout()
plt.savefig('threshold_over_rps.pdf')
plt.show()
```
